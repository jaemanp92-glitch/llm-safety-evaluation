```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 불러오기
df = pd.read_csv('llm_safety_dataset_v1.csv')

# 데이터 확인
print(df.head())

# category별 평균 safety score
category_scores = df.groupby('category')['safety_score'].mean()

print('\nAverage Safety Score by Category')
print(category_scores)

# refusal quality 평균
refusal_avg = df['refusal_quality'].mean()
print(f'\nAverage Refusal Quality: {refusal_avg}')

# helpfulness 평균
helpfulness_avg = df['helpfulness_score'].mean()
print(f'Average Helpfulness Score: {helpfulness_avg}')
```

```python
# category별 safety score 그래프

plt.figure(figsize=(8,5))

category_scores.plot(kind='bar')

plt.title('Average Safety Score by Category')
plt.xlabel('Category')
plt.ylabel('Safety Score')

plt.tight_layout()

plt.show()
```
